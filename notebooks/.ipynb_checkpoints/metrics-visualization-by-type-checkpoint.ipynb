```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# list of models
# Commented few models because they produced very big results which interfere visualization
models = [#'RandomForestRegressor',
          #'AdaBoostRegressor',
          #'BaggingRegressor',
          #'DecisionTreeRegressor',
          'DummyRegressor',
          #'ExtraTreeRegressor',
          #'ExtraTreesRegressor',
          #'GaussianProcessRegressor',
          #'GradientBoostingRegressor',
          #'HuberRegressor',
          #'KNeighborsRegressor',
          #'MLPRegressor',
          #'PassiveAggressiveRegressor',
          #'RANSACRegressor',
          #'SGDRegressor',
          #'TheilSenRegressor'
         ]

dataframes = []
buildingtypes = ['Office', 'PrimClass', 'UnivClass', 'UnivDorm', 'UnivLab']
for model in models:
    for buildingtype in buildingtypes:
        data = pd.read_csv('../results/' + model + '_metrics_' + buildingtype + '.csv')
        data = data.drop(columns=['Unnamed: 0'])
        data['buidingtype'] = buildingtype
        dataframes.append(data)
        result = pd.concat(dataframes)
```

```python
result.head()
```

```python
offices =  result[result['buidingtype']=="Office"]
primclasses =  result[result['buidingtype']=="PrimClass"]
```

```python
import matplotlib.pyplot as plt

print('Model name')

plt.figure(figsize=(15,3))
plt.title('offices', loc='left')
plt.plot(offices['MAPE'])
plt.ylabel('MAPE')
plt.show()


plt.figure(figsize=(15,3))
plt.title('primclasses', loc='left')
plt.plot(primclasses['MAPE'])
plt.ylabel('MAPE')
plt.show()
```
