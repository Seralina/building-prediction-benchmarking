```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize

models = ['AdaBoostRegressor_metrics.csv',
          'BaggingRegressor_metrics.csv',
          'DecisionTreeRegressor_metrics.csv',
          'DummyRegressor_metrics.csv',
          'ExtraTreeRegressor_metrics.csv',
          #'ExtraTreesRegressor_metrics.csv',
          #'GaussianProcessRegressor_metrics.csv',
          #'GradientBoostingRegressor_metrics.csv',
          #'HuberRegressor_metrics.csv',
          #'KNeighborsRegressor_metrics.csv',
          #'MLPRegressor_metrics.csv',
          #'PassiveAggressiveRegressor_metrics.csv',
          #'RANSACRegressor_metrics.csv',
          #'SGDRegressor_metrics.csv',
          #'TheilSenRegressor_metrics.csv'
         ]

dataFrames = []
keys = []
for model in models:
    data = pd.read_csv(model, index_col='Unnamed: 0')
    dataFrames.append(data)
    keys.append(model )
```

```python
# Generate different line styles
lineStyles = ['-', '--', '-.', ':']
lineColors = ['b', 'g', 'r', 'c']
styles = []

for j in range(3):
    for i in range(4):
        styles.append(lineColors[i] + lineStyles[(i + j) % 4])
```

```python
dataList = []
for elem in dataFrames:
    dataList.append(elem['MAPE'].values)
    print(len(elem['MAPE']))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")
```

```python
values = []
for i in range(len(dataList)):
    values.append(dataList[i])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(40, 7.195))
import numpy as np
```

```python
# Line creator
def createLine(values, styleIndex, label):
    return plt.plot(values, styles[styleIndex], label=label)
```

```python
lines = []
for i in range(len(keys)):
    value, = createLine(values[i], i, keys[i])
    lines.append(value)
    

plt.legend(handles=lines)
```
