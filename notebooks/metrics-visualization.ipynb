# Visualization of Model Fit Metrics

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# list of models
# Commented few models because they produced very big results which interfere visualization
models = [#'RandomForestRegressor',
          #'AdaBoostRegressor',
          #'BaggingRegressor',
          #'DecisionTreeRegressor',
          'DummyRegressor',
          'ExtraTreeRegressor',
          #'ExtraTreesRegressor',
          #'GaussianProcessRegressor',
          #'GradientBoostingRegressor',
          #'HuberRegressor',
          'KNeighborsRegressor',
          #'MLPRegressor',
          #'PassiveAggressiveRegressor',
          #'RANSACRegressor',
          #'SGDRegressor',
          #'TheilSenRegressor'
         ]

dataFrames = []
#buildingtypes = ['Office', 'PrimClass', 'UnivClass', 'UnivDorm', 'UnivLab']
for model in models:
    data = pd.read_csv('../results/' + model + '_metrics_PrimClass.csv', index_col='Unnamed: 0')
    dataFrames.append(data)
    
```

```python
# Generate different line styles
# 24 different different lines will be generated
lineStyles = ['-', '--', '-.', ':']
lineColors = ['b', 'g', 'r', 'c', 'm', 'y']
styles = []

for j in range(3):
    for i in range(5):
        styles.append(lineColors[i] + lineStyles[(i + j) % 4])
```

```python
# MAPE,NMBE,CVRSME,RSQUARED possible arguments
# Creates visualization based on argument
def visualize(arg):
    dataList = []
    values = []
    for elem in dataFrames:
        dataList.append(elem[arg].values)
    
    for i in range(len(dataList)):
        values.append(dataList[i])
        
    plt.style.use('seaborn-whitegrid')
    plt.figure(figsize=(15,10))
    
    # Draw lines on single plot
    for i in range(len(models)):
        # Single line creator
        value, = plt.plot(values[i], styles[i], label=models[i])
    
    # Draw plot
    plt.xlabel('Building')
    plt.ylabel(arg + ' value')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()
```

```python
visualize('NMBE')
```

```python
visualize('MAPE')
```

```python
visualize('CVRSME')
```

```python
visualize('RSQUARED')
```
